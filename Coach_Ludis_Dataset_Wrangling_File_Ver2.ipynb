### Code for dealing with "Table S4: Hospitalised football injury cases and population-based rates by age and sex, Australia, 2016–17"

```python
# Readind data into dataframe

injury_allsport_count_df = read_excel(file_name, sheet_name = sheet_name3)
injury_allsport_count_df
```

#### Creating separate dataframes for each sport 

```python
#_________FOR AUSTRALIAN RULES__________


injury_Australianrules_count_df = injury_allsport_count_df.iloc[4:10, :]


# To change the column names
injury_Australianrules_count_df = injury_Australianrules_count_df.reset_index(drop=True)
#injury_allsport_count_df.columns = ['Age Group', 'Males_Count', 'Males(per 100,000 population)', '3', 'Females_Count','Females(per 100,000 population)','6','Persons_Count','Persons(per 100,000 population)']

# Dropping all empty columns
#injury_allsport_count_df = injury_allsport_count_df.drop(['3', '6'], axis=1) #IMPORTANT COMMAND

injury_Australianrules_count_df
```

```python
#_________FOR RUGBY__________


injury_Rugby_count_df = injury_allsport_count_df.iloc[12:18, :]

injury_Rugby_count_df = injury_Rugby_count_df.reset_index(drop=True)

injury_Rugby_count_df
```

```python
#_________FOR SOCCER__________


injury_Soccer_count_df = injury_allsport_count_df.iloc[20:26, :]

injury_Soccer_count_df = injury_Soccer_count_df.reset_index(drop=True)

injury_Soccer_count_df
```

```python
#_________FOR TOUCH FOOTBALL__________


injury_TochFootball_count_df = injury_allsport_count_df.iloc[28:34, :]

injury_TochFootball_count_df = injury_TochFootball_count_df.reset_index(drop=True)

injury_TochFootball_count_df

```

```python
#_________FOR OTHER AND UNSPECIFIED FOOTBALL____________________



injury_OtherFootball_count_df = injury_allsport_count_df.iloc[36:42, :]

injury_OtherFootball_count_df = injury_OtherFootball_count_df.reset_index(drop=True)

injury_OtherFootball_count_df

```

```python
PersonsCount_AustralianRules = injury_Australianrules_count_df.at[0,'Persons_Count']
print ("PersonsCount_AustralianRules :", PersonsCount_AustralianRules)

PersonsCount_Rugby = injury_Rugby_count_df.at[0,'Persons_Count']
print ("PersonsCount_Rugby", PersonsCount_Rugby)

PersonsCount_Soccer = injury_Soccer_count_df.at[0,'Persons_Count']
print ("PersonsCount_Soccer", PersonsCount_Soccer)

PersonsCount_TouchFootball = injury_TochFootball_count_df.at[0,'Persons_Count']
print ("PersonsCount_TouchFootball", PersonsCount_TouchFootball)

PersonsCount_Other = injury_OtherFootball_count_df.at[0,'Persons_Count']
print ("PersonsCount_Other",PersonsCount_Other)

totalHospitalized = PersonsCount_AustralianRules + PersonsCount_Rugby + PersonsCount_Soccer + PersonsCount_TouchFootball + PersonsCount_Other
print ("The total number of hospitalizations for the age group 0-14 is :" ,totalHospitalized)
```
