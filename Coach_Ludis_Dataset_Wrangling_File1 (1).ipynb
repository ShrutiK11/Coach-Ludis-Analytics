```python
# Importing Libraries

# Pandas offers data structures and operations for manipulating numerical tables and time series and for data manipulation and analysis
import pandas as pd

#To read .xlxs files
from pandas import read_excel

#import numpy as np
#import re
```

```python
# Loading data using pandas library

# Initializing Sheet name var. All the changes made in the sheet name in the original dataset can be changed here.
# No need to change elsewhere in the code
sheet_name1 = 'BodyPart_InjuryType'
sheet_name2 = 'Mechanism_restALL'
#sheet_name3 = 'Count_Hospitalized_Injury'


# If the .xlsx file name changes, replace 'InjuryData.xlsx' wil 'New_File_Name.xlsx'
file_name = 'InjuryData.xlsx' 
```

```python
# Readind data into dataframe

df = read_excel(file_name, sheet_name = sheet_name1)
#df

mech_df = read_excel(file_name, sheet_name = sheet_name2)
#mech_df
```

```python
#____FOR "BODY PART INJURED"____


# Selecting rows from index 3 to 12
df1 = df.iloc[3:12, :]


# To change the column names
df1 = df1.reset_index(drop=True)
df1.columns = ['Body Part Injured', 'Australian rules_Count', 'Australian rules_Percentage', '3', 'Rugby_Count','Rugby_Percentage','6','Soccer_Count','Soccer_Percentage','9', 'Touch football_Count','Touch football_Percentage','12','Other & Unspecified football_Count','Other & Unspecified football_Percentage','15','Total_Count','Total_Percentage']


# Dropping all empty columns
df1 = df1.drop(['3', '6', '9' , '12', '15'], axis=1) #IMPORTANT COMMAND


# Print the dataframe
df1
```

```python
#____FOR "TYPE OF INJURY"____


# Selecting rows from index 14 to 26
df2 = df.iloc[14:26, :]


# To change the column names
df2 = df2.reset_index(drop=True)
df2.columns = ['Type of Injury', 'Australian rules_Count', 'Australian rules_Percentage', '3', 'Rugby_Count','Rugby_Percentage','6','Soccer_Count','Soccer_Percentage','9', 'Touch football_Count','Touch football_Percentage','12','Other & Unspecified football_Count','Other & Unspecified football_Percentage','15','Total_Count','Total_Percentage']


# Dropping all empty columns
df2 = df2.drop(['3', '6', '9' , '12', '15'], axis=1) #IMPORTANT COMMAND


# Print the dataframe
df2
```

```python
#____FOR "MECHANISM OF INJURY FOR RUGBY, SOCCER, TOUCH FOOTBALL"____

# Selecting rows from index 2 to 8
mech_df = mech_df.iloc[2:8, :]
 
    
# To change the column names
mech_df = mech_df.reset_index(drop=True)
mech_df.columns = ['Mechanism Of Injury', 'Australian rules_Count', 'Australian rules_Percentage', '3', 'Rugby_Count','Rugby_Percentage','6','Soccer_Count','Soccer_Percentage','9', 'Touch football_Count','Touch football_Percentage','12','Other & Unspecified football_Count','Other & Unspecified football_Percentage','15','Total_Count','Total_Percentage']


# Dropping all empty columns
mech_df = mech_df.drop(['3', '6', '9' , '12', '15'], axis=1) #IMPORTANT COMMAND


# Print the dataframe
mech_df


```
